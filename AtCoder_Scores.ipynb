```python
# ABC054 C - One-stroke Path
# 無向グラフ、隣接グラフ、隣接行列
import itertools

N,M = map(int,input().split())

gr = [[0]*N for _ in range(N)]
for i in range(M):
    a,b = map(int,input().split())
    a -= 1
    b -= 1
    gr[a][b] = 1
    gr[b][a] = 1
#print(gr)

def one_stroke(arr):
    st = 0
    for i in range(len(arr)):
        nx = arr[i]
        if gr[st][nx] == 0:
            return False
        st = nx
    return True

ans = 0
for arr in itertools.permutations(range(1,N)):
    #print(arr)
    if one_stroke(arr):
        ans += 1
print(ans)
```
